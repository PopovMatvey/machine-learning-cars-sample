```python
import csv
import pandas as pd
import statsmodels.api as sm

nameFileDataSet = 'data/cars_dataset.csv'
```

```python
# Вывести массив
def printArray(parArray):
    for i in parArray:
        print(i)

# Вывести содержимое csv файла
def printSCVFile(parNameFile):
    with open(parNameFile) as scvFile:
        reader = csv.reader(scvFile)
        for row in reader:
            print(row[0])

#Получить размер выборки (scv)
def getSizeDataSet(parNameFile):
    with open(parNameFile) as scvFile:
        reader = csv.reader(scvFile)
        return sum(1 for row in reader)
    
#
def getArrayElementProperty(parNameFile,parIndexProperty):
    returnedArray = []
    index = 0
    with open(parNameFile) as scvFile:
        reader = csv.reader(scvFile)
        for row in reader:
            if(index != 0):
                returnedArray.append(row[parIndexProperty])
            index += 1
    return returnedArray

#
def getAmountProperty(parNameFile):
    with open(parNameFile) as scvFile:
        reader = csv.reader(scvFile)
        for row in reader:
            return len(row)

#Получить заголовки
def getHeader(parNameFile):
    with open(parNameFile) as scvFile:
        reader = csv.reader(scvFile)
        for row in reader:
            return row

        
# Получить объект Pandas (Автомобили)
def getObjectPDCars(parNameFile):
    return {
        getHeader(parNameFile)[0]:getArrayElementProperty(parNameFile,0),
        getHeader(parNameFile)[1]:getArrayElementProperty(parNameFile,1),
        getHeader(parNameFile)[2]:getArrayElementProperty(parNameFile,2),
        getHeader(parNameFile)[3]:getArrayElementProperty(parNameFile,3),
        getHeader(parNameFile)[4]:getArrayElementProperty(parNameFile,4),
        getHeader(parNameFile)[5]:getArrayElementProperty(parNameFile,5),
        getHeader(parNameFile)[6]:getArrayElementProperty(parNameFile,6),
        getHeader(parNameFile)[7]:getArrayElementProperty(parNameFile,7),
        getHeader(parNameFile)[8]:getArrayElementProperty(parNameFile,8),
        getHeader(parNameFile)[9]:getArrayElementProperty(parNameFile,9),
        getHeader(parNameFile)[10]:getArrayElementProperty(parNameFile,10),
        getHeader(parNameFile)[11]:getArrayElementProperty(parNameFile,11),
        getHeader(parNameFile)[12]:getArrayElementProperty(parNameFile,12),
        getHeader(parNameFile)[13]:getArrayElementProperty(parNameFile,13),
        getHeader(parNameFile)[14]:getArrayElementProperty(parNameFile,14),
        getHeader(parNameFile)[15]:getArrayElementProperty(parNameFile,15),
        getHeader(parNameFile)[16]:getArrayElementProperty(parNameFile,16),
        getHeader(parNameFile)[17]:getArrayElementProperty(parNameFile,17),
        getHeader(parNameFile)[18]:getArrayElementProperty(parNameFile,18),
        getHeader(parNameFile)[19]:getArrayElementProperty(parNameFile,19),
        getHeader(parNameFile)[20]:getArrayElementProperty(parNameFile,20),
        getHeader(parNameFile)[21]:getArrayElementProperty(parNameFile,21),
        getHeader(parNameFile)[22]:getArrayElementProperty(parNameFile,22),
        getHeader(parNameFile)[23]:getArrayElementProperty(parNameFile,23),
        getHeader(parNameFile)[24]:getArrayElementProperty(parNameFile,24),
    }

# 
def changeArrayToFloat(parArray):
    returnedArray = []
    for item in parArray:
        returnedArray.append(float(item))
    return returnedArray
            
```
